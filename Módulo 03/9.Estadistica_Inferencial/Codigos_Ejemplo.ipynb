# Código para la Distribución de la Muestra

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generando datos de muestra
np.random.seed(0)
data = np.random.normal(loc=0, scale=1, size=100)

# Creación del histograma
plt.figure(figsize=(8, 6))
sns.histplot(data, bins=10, kde=True, color="blue")
plt.title('Distribución de la Muestra')
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.show()
```

# Código para Hipótesis Nula vs Hipótesis Alternativa

```python
plt.figure(figsize=(8, 6))

# Distribución normal bajo H0
x = np.linspace(-4, 4, 1000)
y = (1/np.sqrt(2*np.pi)) * np.exp(-0.5*x**2)
plt.plot(x, y, label='Distribución bajo H0', color='blue')

# Distribución normal bajo H1
y_alt = (1/np.sqrt(2*np.pi)) * np.exp(-0.5*(x-2)**2)
plt.plot(x, y_alt, label='Distribución bajo H1', color='red')

plt.fill_between(x, y, 0, where=(x>1.96), color='blue', alpha=0.3, label='Región de rechazo (H0)')
plt.fill_between(x, y_alt, 0, where=(x>1.96), color='red', alpha=0.3, label='Región de aceptación (H1)')

plt.title('Hipótesis Nula vs Hipótesis Alternativa')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

# Código para Intervalos de Confianza

```python
np.random.seed(42)
sample_means = [np.mean(np.random.normal(loc=0, scale=1, size=30)) for _ in range(100)]
ci_lower = np.percentile(sample_means, 2.5)
ci_upper = np.percentile(sample_means, 97.5)

plt.figure(figsize=(8, 6))
sns.histplot(sample_means, kde=True, color="purple")
plt.axvline(ci_lower, color='red', linestyle='--', label=f'CI Lower ({ci_lower:.2f})')
plt.axvline(ci_upper, color='red', linestyle='--', label=f'CI Upper ({ci_upper:.2f})')
plt.axvline(np.mean(sample_means), color='black', linestyle='-', label=f'Mean ({np.mean(sample_means):.2f})')
plt.title('Intervalos de Confianza')
plt.xlabel('Media Muestral')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

# Código para Errores Tipo I y Tipo II

```python
plt.figure(figsize=(8, 6))

# Distribución normal bajo H0
x = np.linspace(-4, 4, 1000)
y = (1/np.sqrt(2*np.pi)) * np.exp(-0.5*x**2)
plt.plot(x, y, label='Distribución bajo H0', color='blue')

# Distribución normal bajo H1
y_alt = (1/np.sqrt(2*np.pi)) * np.exp(-0.5*(x-2)**2)
plt.plot(x, y_alt, label='Distribución bajo H1', color='red')

plt.fill_between(x, y, 0, where=(x>1.96), color='blue', alpha=0.3, label='Error Tipo I')
plt.fill_between(x, y_alt, 0, where=(x<=1.96), color='red', alpha=0.3, label='Error Tipo II')

plt.title('Errores Tipo I y Tipo II')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

```python
np.random.seed(0)
sample_means = [np.mean(np.random.normal(loc=0, scale=1, size=30)) for _ in range(1000)]

plt.figure(figsize=(8, 6))
sns.histplot(sample_means, kde=True, color="green")
plt.title('Distribución Muestral de la Media')
plt.xlabel('Media Muestral')
plt.ylabel('Frecuencia')
plt.show()
```

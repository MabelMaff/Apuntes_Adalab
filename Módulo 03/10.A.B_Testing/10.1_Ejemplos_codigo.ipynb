# Gráfico de Distribución de Tasa de Conversión

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generar datos de ejemplo
np.random.seed(0)
conversion_control = np.random.normal(loc=0.05, scale=0.01, size=1000)
conversion_test = np.random.normal(loc=0.06, scale=0.01, size=1000)

# Visualizar la distribución de la tasa de conversión
plt.figure(figsize=(10, 6))
sns.histplot(conversion_control, kde=True, color='blue', label='Control')
sns.histplot(conversion_test, kde=True, color='green', label='Test')
plt.title('Distribución de la Tasa de Conversión')
plt.xlabel('Tasa de Conversión')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```

# Gráfico de Comparación de Medias

```python
import numpy as np
import matplotlib.pyplot as plt

# Datos de ejemplo
np.random.seed(0)
conversion_control = np.random.normal(loc=0.05, scale=0.01, size=1000)
conversion_test = np.random.normal(loc=0.06, scale=0.01, size=1000)

# Calcular medias y desviaciones estándar
media_control = np.mean(conversion_control)
media_test = np.mean(conversion_test)
std_control = np.std(conversion_control)
std_test = np.std(conversion_test)

# Visualizar la comparación de medias
plt.figure(figsize=(10, 6))
labels = ['Control', 'Test']
medias = [media_control, media_test]
stds = [std_control, std_test]

plt.bar(labels, medias, yerr=stds, color=['blue', 'green'], capsize=10)
plt.title('Comparación de Medias de Tasa de Conversión')
plt.ylabel('Tasa de Conversión')
plt.show()
```

# Gráfico de Resultados del A/B Testing

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generar datos de ejemplo
np.random.seed(0)
conversion_control = np.random.normal(loc=0.05, scale=0.01, size=1000)
conversion_test = np.random.normal(loc=0.06, scale=0.01, size=1000)

# Crear un DataFrame para los resultados
import pandas as pd
data = {
    'Grupo': ['Control'] * 1000 + ['Test'] * 1000,
    'Tasa de Conversión': np.concatenate([conversion_control, conversion_test])
}
df = pd.DataFrame(data)

# Visualizar los resultados del A/B Testing
plt.figure(figsize=(10, 6))
sns.boxplot(x='Grupo', y='Tasa de Conversión', data=df)
plt.title('Resultados del A/B Testing')
plt.ylabel('Tasa de Conversión')
plt.show()
```

# Visualización de la Prueba t de Student

```python
import numpy as np
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt

# Generar datos de ejemplo
np.random.seed(0)
conversion_control = np.random.normal(loc=0.05, scale=0.01, size=1000)
conversion_test = np.random.normal(loc=0.06, scale=0.01, size=1000)

# Realizar la prueba t de Student
t_stat, p_value = ttest_ind(conversion_control, conversion_test)

# Visualizar la distribución de los datos y el resultado de la prueba t
plt.figure(figsize=(10, 6))
sns.histplot(conversion_control, kde=True, color='blue', label='Control')
sns.histplot(conversion_test, kde=True, color='green', label='Test')
plt.axvline(np.mean(conversion_control), color='blue', linestyle='--', label=f'Media Control: {np.mean(conversion_control):.4f}')
plt.axvline(np.mean(conversion_test), color='green', linestyle='--', label=f'Media Test: {np.mean(conversion_test):.4f}')
plt.title(f'Resultado de la Prueba t de Student (t={t_stat:.4f}, p={p_value:.4f})')
plt.xlabel('Tasa de Conversión')
plt.ylabel('Frecuencia')
plt.legend()
plt.show()
```
